# Exploring EBI gene2pheno data for parser

## Downloading data

Using the "all data" option/file. 

Latest data (created on-the-fly) can be downloaded from:
* API urls https://www.ebi.ac.uk/gene2phenotype/api/: `panel/{name}/download/` endpoint. Has "all data" option! Works programmatically and manually
* webpage https://www.ebi.ac.uk/gene2phenotype/download: currently manual clicks only. Has "all data" option. 

Static release files can be downloaded from the FTP site http://ftp.ebi.ac.uk/pub/databases/gene2phenotype/ (in the G2P_data_downloads subfolder). These releases included the "all data" file starting in Aug 2025. It updates on a monthly basis, according to the [README](http://ftp.ebi.ac.uk/pub/databases/gene2phenotype/README). However, I notice that the FTP site is sometimes slow on my computer (Firefox browser).

## Notes on parsing data to create documents 

Reordered for what makes sense to put together. <mark>Highlight</mark> means this column is important for ingest

### Gene subject section

<mark>**hgnc id**</mark>: single ID, no prefix. No NA so easier to use (need to assign dtype to be string, otherwise may default load as int)

---

**gene mim** (a few NA): OMIM ID for gene (single ID), no prefix. NodeNorm does recognize these (need to assign dtype to be string, otherwise may default load as float)

**gene symbol**: str. "HGNC-assigned gene symbol" (according to Data_download_format txt file)

**previous gene symbols**: "; "-delimited. 

### Disease object section

<mark>**disease mim**</mark> (some NA): OMIM ID for disease ("or highly similar" according to Data_download_format txt file). Single ID, no prefix. Used to find orphanet IDs mixed in (prefix `Orphanet:`) but haven't seen as of 2025-12. (need to assign dtype to be string, otherwise may default load as int)

<mark>**disease MONDO**</mark> (some NA): MONDO ID for disease ("or highly similar" according to Data_download_format txt file). Already Translator-formatted CURIE

**disease name**: Previously reviewed and didn't find any IDs, did find potential delimiters (`:`, `;`). Previously found surrounding whitespace to strip

### Association section

---

**Content**

<mark>**g2p id**</mark>: resource's unique stable ID for each record/row (according to Data_download_format txt file, G2P webpage for single record) => use to generate links to edge info. Current format: https://www.ebi.ac.uk/gene2phenotype/lgd/G2P03386


<mark>**allelic requirement**</mark>: required genotype (according to G2P webpage for single record). str (categorical). [G2P website](https://www.ebi.ac.uk/gene2phenotype/about/terminology) says these are synonyms of HPO "mode of inheritance terms", which seem to be from the [inheritance part of the ontology](https://hpo.jax.org/browse/term/HP:0000005). I've included a mapping table in reference section.

<mark>**confidence**</mark>: confidence that the association is real (according to G2P webpage for panel). [G2P website](https://www.ebi.ac.uk/gene2phenotype/about/terminology) has definitions of terms. str (categorical)
- OLD biolink note: there's an association-slot `has confidence level`. But there's also an entity `confidence level` that's supposed to have values from CIO. Are G2P's terms okay, or are they supposed to be mapped to ontology terms like CIO/SEPIO? (which...may be a loss of info compared to G2P's term definitions)

<mark>**molecular mechanism**</mark>: mechanism of how gene's variants causes disease (according to G2P webpage for single record). str (categorical). [G2P website](https://www.ebi.ac.uk/gene2phenotype/about/terminology) has definitions of terms

<mark>**publications**</mark> (some NA): reviewed to create this row/association (according to Data_download_format file). These are PMIDs with no prefix (based on G2P webpage for single record). Can be "; "-delimited

<mark>**date of last review**</mark>: date the record was modified or reviewed (according to Data_download_format txt file). In single-record webpage, this is called "Last Updated On".

---

**cross cutting modifier** (lots of NA!): additional info relevant to gene-disease inheritance (according to G2P webpage for single record). Can be "; "-delimited, otherwise would be categorical. [G2P website](https://www.ebi.ac.uk/gene2phenotype/about/terminology#cross-cutting-modifier-section) has definitions of terms and says these are HPO inheritance qualifier terms "when available" ([this part of ontology](https://hpo.jax.org/browse/term/HP:0034335)). But I previously found some terms are custom. I didn't try mapping these.

<br>

**variant consequence** (some NA): consequences of reported variants on product (protein or RNA) (according to G2P webpage for single record). Can be "; "-delimited, otherwise would be categorical. [G2P website](https://www.ebi.ac.uk/gene2phenotype/about/terminology) has definitions of terms and SO mappings. 

**variant types** (lots of NA!): associated with gene-disease pair (according to G2P webpage for single record). Can be "; "-delimited, otherwise would be categorical. [G2P website](https://www.ebi.ac.uk/gene2phenotype/about/terminology) has definitions of terms and SO mappings (large table!). **>100 UNIQUE VALUES (SINGLE, MIX OF TERMS). I don't think using this as a qualifier is a good idea - it's confusing**
* OLD biolink note: this info kinda exists in biolink as predicates for variant-gene relationships.

<br>

**molecular mechanism support** (new Aug 2025): seems to say how molecular mechanism was decided. But not sure exactly what this means/what methods - maybe in paper? str (categorical): "inferred" or "evidence".
"inferred" does show up on record webpage. Ex: "Molecular mechanism" > "Mechanism" will say "loss of function (inferred)".
Unclear how "evidence" alone would show up on webpage - rows always have "molecular mechanism evidence" value that shows up as "Molecular mechanism" > "Evidence" (current ex:  https://www.ebi.ac.uk/gene2phenotype/lgd/G2P00018)

**molecular mechanism categorisation** (new Aug 2025; **VERY SPARSE**): more detailed mechanism info, structured (displayed in single-record webpage as a table). 
- key: term from the ["molecular mechanism synopsis" section of terminology page](https://www.ebi.ac.uk/gene2phenotype/about/terminology#mechanism-synopsis-section). Separator between key and value is ":" 
- value: either 'evidence' or 'inferred' (no missing values, usually "evidence"). 
- Can have multiple key-value pairs ("; "-delimited). Each key-value pair becomes one row in molecular mechanism -> categorisation section of single-record webpage. 

**molecular mechanism evidence** (**VERY SPARSE**): "types of evidence available to support reported mechanism", according to Data download format txt. 
Has value in all rows where "molecular mechanism support" == "evidence" and a few rows where == "inferred".
complicated structure, is displayed in single-record webpages as a table under "Molecular mechanism" > "Evidence" (current ex: https://www.ebi.ac.uk/gene2phenotype/lgd/G2P00064)
- `publication_ID -> functional studies` (delimiter is " -> "). Functional studies key:value pairs are described in ["molecular mechanism evidence types" section of terminology page](https://www.ebi.ac.uk/gene2phenotype/about/terminology#mechanism-evidence-section).
  - key: class (enum). Separator between key and value is ":".
  - value: "evidence type". Possible value depends on key.
  - can have multiple key-value pairs ("; "-delimited) for 1 publication ID. Becomes different bullet points in row 
- can have info from multiple publications (" & "-delimited). Become different rows

<br>

**additional mined publications** (some NA): mined from PubMed but not reviewed yet (according to Data_download_format file). These are PMIDs with no prefix (based on G2P webpage for single record). Can be "; "-delimited

<br>

**phenotypes** (some NA): HPO IDs (already Translator-formatted CURIE) for phenotypes reported by the publications (according to G2P webpage for single record). Can be "; "-delimited
- based on G2P website, these are organized/assigned to specific gene-disease records...so I'm keeping as a part of the gene-disease association, rather than creating diff objects for gene-phenotype (if I did this, it would be a separate function)

<br>

**panel**: the G2P panels this record is assigned to (according to Data_download_format txt file). Can be "; "-delimited, otherwise would be categorical.
- map (str.replace) values to match G2P webpage for single record
  - "DD" = "Developmental disorders"
  - add " disorders" to end: "Cancer", "Cardiac", "Eye", "Skeletal", "Skin"
  - keep as-is: "Hearing loss"
    
    
**comments** (**VERY SPARSE**): additional comments from curation team (according to Data_download_format txt file). Appears to be free-text (sometimes short, sometimes very long). Sometimes has surrounding whitespace to strip 

**review** (new Aug 2025): indicates when record is under review (according to Data_download_format txt file). Don't know what the possible values are (boolean?) because I've only seen the column be all NA so far.
